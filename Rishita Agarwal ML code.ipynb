```python
df = pd.get_dummies(df,columns = ["ane"]) # Change the object value into numerical datatype.
df = pd.get_dummies(df,columns = ["pe"])
df = pd.get_dummies(df,columns = ["appet"])
df = pd.get_dummies(df,columns = ["cad"])
df = pd.get_dummies(df,columns = ["dm"])
df = pd.get_dummies(df,columns = ["htn"])
df = pd.get_dummies(df,columns = ["ba"])
df = pd.get_dummies(df,columns = ["pcc"])
df = pd.get_dummies(df,columns = ["pc"])
df = pd.get_dummies(df,columns = ["rbc"]) # Change the object value into numerical datatype.
```

```python
df.drop(['rc', 'pcv',"wc"], axis = 1)
```

```python
x = df.drop(["classification"],axis  =1) # It contains the whole dataset except the price column
y = df[["classification"]] # It contains the price column
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 2) # Dataset distributed in 70:30
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression() # Initiated the model here
regression_model.fit(x_train,y_train) # Train this data using linear regression
```

```python
regression_model.score(x_train,y_train) # 95% accuracy in training data using linear regression
```

```python
regression_model.score(x_test,y_test) # 75% accuracy in testing the data
```

```python
y_predict = regression_model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.ensemble import RandomForestRegressor
regression_model1 = RandomForestRegressor(n_estimators = 100, random_state = 1) # Initialisation
regression_model1.fit(x_train,y_train) # fitting this Regression to train data
```

```python
regression_model1.score(x_train,y_train) # 98% accuracy in training data using Random Forest Regression
```

```python
regression_model1.score(x_test,y_test) # 86% accuracy in testing data using Random Forest Regression
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
score_1 = []
for i in range (1,50):
  KNN_2 = KNeighborsClassifier(n_neighbors = i)
  score_2 = cross_val_score(KNN_2,x,y,cv = 10) #cv=10 means repeat the iteration 10 times
  score_1.append(score_2.mean())
```
